```python
import matplotlib.pyplot as plt
import pyplotbrookings as ppb
import seaborn as sns
```

```python
#Brookings plot theme
ppb.set_theme()
ppb.set_pallet('brand2', reverse=False)

penguins = sns.load_dataset("penguins")
plt.hist(penguins[penguins['species'] == 'Adelie']['bill_depth_mm'], alpha=0.5, label='Adeline')
plt.hist(penguins[penguins['species'] == 'Chinstrap']['bill_depth_mm'], alpha=0.5, label='Chinstrap')
plt.hist(penguins[penguins['species'] == 'Gentoo']['bill_depth_mm'], alpha=0.5, label='Gentoo')

plt.legend(ncol=3)
plt.xlabel('Bill Depth (mm)')
plt.ylabel('Count')

# Adding Brookings titles and logos
ppb.add_title(title='Penguin Bill Length', subtitle='Bill length in mm of Adelie and Gentoo Penguins at Palmer Station LTER', 
          tag='FIGURE 1', source='Palmer Station Antarctica LTER', notes='Figure made using matplotlib')
ppb.add_logo('hc')
```
